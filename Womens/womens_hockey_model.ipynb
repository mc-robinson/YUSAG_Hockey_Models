```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import math
```

```python
sixteen_df = pd.read_csv('./uscho_games_2016.csv',header=None)
sixteen_df['season'] = pd.Series(2016, index=sixteen_df.index)
seventeen_df = pd.read_csv('./uscho_games_2017.csv',header=None)
seventeen_df['season'] = pd.Series(2017, index=seventeen_df.index)
eighteen_df = pd.read_csv('./uscho_games_2018.csv',header=None)
eighteen_df['season'] = pd.Series(2018, index=eighteen_df.index)
eighteen_df = eighteen_df.dropna(subset=[9])

df = pd.concat([sixteen_df,seventeen_df,eighteen_df],ignore_index=True)
```

```python
df.head()
```

```python
df = df[[8,9,14,15,'season']]
```

```python
df.head()
```

```python
df = df.rename(columns={8: 'team', 9: 'teamscore', 14: 'opponent', 15: 'oppscore'})
```

```python
df.head()
```

```python
df['weights'] = (df['season']-2012)
df['weights'] = df['weights'].apply(lambda x: math.factorial(x))
```

```python
df.head()
```

```python
df['location'] = 'V'
```

```python
df.head()
```

```python
import re
```

```python
df.head()
```

```python
df['team'] = df['team'].apply(lambda x: re.sub("[\(\[].*?[\)\]]", "", x))
df['team'] = df['team'].apply(lambda x: x.lstrip())

df['opponent'] = df['opponent'].apply(lambda x: re.sub("[\(\[].*?[\)\]]", "", x))
df['opponent'] = df['opponent'].apply(lambda x: x.lstrip())
```

```python
df.head()
```

```python
df['scorediff'] = (df['teamscore']-df['oppscore'])
```

```python
df['location'] = df['location'].replace('V',-1)
df['location'] = df['location'].replace('N',0)
df['location'] = df['location'].replace('H',1)
```

```python
df.head()
```

```python
df1 = df.copy()
df1.head()
```

```python
df1 = df1.rename(columns={'team': 'opponent', 'teamscore': 'oppscore', 'opponent': 'team', 'oppscore':'teamscore'})
```

```python
df1.head()
```

```python
df1['scorediff'] = df1['scorediff'].apply(lambda x: x*-1)
df1['location'] = df1['location'].apply(lambda x: x*-1)
```

```python
df1.head()
```

```python
df.shape
```

```python
df1.shape
```

```python
df = pd.concat([df,df1],ignore_index=True)
```

```python
df.head()
```

```python
df = df[~df.team.str.contains("/")]
df = df[~df.opponent.str.contains("/")]
```

```python
# create dummy variables, need to do this in python b/c does not handle automatically like R
team_dummies = pd.get_dummies(df.team, prefix='team')
opponent_dummies = pd.get_dummies(df.opponent, prefix='opponent')

df = pd.concat([df, team_dummies, opponent_dummies], axis=1)
```

```python
df.head()
```

```python
df = df.dropna(subset=['scorediff','weights'])
```

```python
# make the training data
X = df.drop(['team','opponent','teamscore','oppscore','season','weights','scorediff'], axis=1)
y = df['scorediff']
weights = df['weights']
```

```python
X.head()
```

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge()
ridge_reg.fit(X, y, sample_weight=weights)
```

```python
# get the R^2 value
r_squared = ridge_reg.score(X, y, sample_weight=weights)
print('R^2 on the training data:')
print(r_squared)
```

```python
# get the coefficients for each feature
coef_data = list(zip(X.columns,ridge_reg.coef_))
coef_df = pd.DataFrame(coef_data,columns=['feature','feature_coef'])
coef_df.head()
```

```python
# first get rid of opponent_ variables
team_df = coef_df[~coef_df['feature'].str.contains("opponent")]

# get rid of the location variable
team_df = team_df.iloc[1:]
```

```python
# rank them by coef, not alphabetical order
ranked_team_df = team_df.sort_values(['feature_coef'],ascending=False)
# reset the indices at 0
ranked_team_df = ranked_team_df.reset_index(drop=True);
ranked_team_df.rename(columns={'feature':'team', 'feature_coef':'YUSAG_coef'}, inplace=True)
ranked_team_df['team'] = ranked_team_df['team'].str.replace('team_', '')
```

```python
ranked_team_df
```

```python
ivy_team_names = ['Yale','Harvard','Princeton','Cornell','Brown','Columbia','Dartmouth','Penn']
ivy_ranked_team_df = ranked_team_df[ranked_team_df['team'].isin(ivy_team_names)]
```

```python
ivy_ranked_team_df
```

```python
ECAC_team_names = ['Yale','Harvard','Princeton','Cornell','Brown','Columbia','Dartmouth','Penn',
                  'Clarkson','Colgate','Union','St. Lawrence','Quinnipiac','Rensselaer']
ECAC_ranked_team_df = ranked_team_df[ranked_team_df['team'].isin(ECAC_team_names)]
```
